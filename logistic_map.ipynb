### Logistic map 

Lite studium av f√∏lgen 

$$
x_{n+1} = r x_n (1 - x_n)
$$

Der $x_n \in [0,1]$

```python
import numpy as np 
import matplotlib.pyplot as plt
 
import time
```

```python
iteration = lambda x, r: r*x*(1-x)
def simulate(X0, R, N):
    X = np.copy(X0)
    for k in range(N):
        X = iteration(X, R)
    return X
```

```python
N = 100
M = 10000
x0 = np.random.uniform(0,1.0,N)
X = x0*np.ones((M,N))
r = np.linspace(2.8,4.00,M)
R =  r*(np.ones((N,M)))
R = R.T

X = simulate(X, R, 2)

plt.figure(figsize = (20,10))
[plt.plot(r, X[:, i] , ',', color = 'blue', alpha = 0.1) for i in range(X.shape[1])]
plt.title(f'Iterations = 2')

X = simulate(X, R, 10)

plt.figure(figsize = (20,10))
[plt.plot(r, X[:, i] , ',', color = 'blue', alpha = 0.1) for i in range(X.shape[1])]
plt.title(f'Iterations = 12')

X = simulate(X, R, 10)

plt.figure(figsize = (20,10))
[plt.plot(r, X[:, i] , ',', color = 'blue', alpha = 0.1) for i in range(X.shape[1])]
plt.title(f'Iterations = 22')

X = simulate(X, R, 10)

plt.figure(figsize = (20,10))
[plt.plot(r, X[:, i] , ',', color = 'blue', alpha = 0.1) for i in range(X.shape[1])]
plt.title(f'Iterations = 32')

X = simulate(X, R, 100)

plt.figure(figsize = (20,10))
[plt.plot(r, X[:, i] , ',', color = 'blue', alpha = 0.1) for i in range(X.shape[1])]
plt.title(f'Iterations = 132')
plt.show()
```
